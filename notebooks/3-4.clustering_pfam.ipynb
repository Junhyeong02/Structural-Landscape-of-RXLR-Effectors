```python
import pandas as pd
from pprint import pprint
```

```python
df = pd.read_csv("total.tsv", sep = "\t")
```

```python
df.fillna("<unk>", inplace = True)
```

```python
cluster_assigned_pfam = {}

for idx in df.index:
    str_cl = int(df.loc[idx, "str_cl"])
    domain = df.loc[idx, "domain"]

    if str_cl == -1:
        continue

    if str_cl not in cluster_assigned_pfam:
        cluster_assigned_pfam[str_cl] = {}

    if domain not in cluster_assigned_pfam[str_cl]:
        cluster_assigned_pfam[str_cl][domain] = 0

    cluster_assigned_pfam[str_cl][domain] += 1
```

```python
WY, LWY, nWY = [], [], []

for key in range(1,30):
    data = cluster_assigned_pfam[key]

    s = sum(val for val in data.values())
    
    LWY.append(data["LWY"] if "LWY" in data else 0)
    WY.append(data["WY"] if "WY" in data else 0)
    nWY.append(s - LWY[-1] - WY[-1])
    
```

```python
results = pd.DataFrame()
results["WY"] = WY
results["LWY"] = LWY
results["nWY"] = nWY
```

```python
results.to_csv("data1.txt", sep = "\t", index = False)
```
