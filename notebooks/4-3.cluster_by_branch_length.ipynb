```python
from Bio import Phylo
import pandas as pd

from glob import glob
```

```python
%cd ../analysis/4.WY_tree
```

```python
tree_cluster = {}
```

```python
for i, hsc in enumerate(sorted(glob("./hSC*.txt"))):
    with open(hsc) as f:
        print(hsc)
        for line in f.readlines():
            tree_cluster[line.strip().replace(" ", "_")] = i+1
```

```python
df = pd.read_csv("./RXLR_WY.tsv",  sep = "\t")
```

```python
df["tree_cluster"] = df["entry"].map(tree_cluster)
```

```python
df_w_cluster = df.sort_values(["tree_cluster", "str_cl", "seq_cl", "entry"])
```

```python
df_w_cluster.to_csv("RXLR_tree_cluster_by_manual.tsv", sep = "\t", index = False)
```
