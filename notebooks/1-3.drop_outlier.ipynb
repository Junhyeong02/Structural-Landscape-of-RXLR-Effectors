```python
import os

from glob import glob
from Bio import SeqIO
```

```python
%cd ../analysis/1.RXLR_prediction/1-3.drop_outlier/
```

```python
fastas = glob("../1-2.pattern_match/results/*_signalp_regex.fasta")
```

```python
for fasta in fastas:
    out = os.path.basename(fasta).replace("_regex", "_regex_length")

    out = os.path.join("results/", out)
    
    with open(out, "w") as f:
        for record in SeqIO.parse(fasta, "fasta"):
            entry = record.id
            seq = str(record.seq)

            if len(seq) > 1500 or len(seq) < 16:
                continue

            f.write(f">{entry}\n{seq}\n")
```
