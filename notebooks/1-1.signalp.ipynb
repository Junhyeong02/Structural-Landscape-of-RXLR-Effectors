```python
import pandas as pd
from Bio import SeqIO
import os
from glob import glob
```

```python
%cd ../analysis/1.RXLR_prediction/1-1.signalp/
```

```python
for fasta in glob("results/*/processed_entries.fasta"):
    spp = fasta.split('/')[1]
    out = f"results.parse/{spp}_signalp.fasta"

    print(out)
    
    with open(out, "w") as f:
        for record in SeqIO.parse(fasta, "fasta"):
            entry = record.id
            if entry.startswith("jgi"):
                temp = entry.split("|")
                entry = temp[1] +"_" + temp[2]
            
            seq = str(record.seq)
            seq = seq.strip("*")
    
            f.write(f">{entry}\n{seq}\n")
```
