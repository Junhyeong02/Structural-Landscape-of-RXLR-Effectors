```python
import pandas as pd
```

```python
pfam = pd.read_csv("../3.Clustering/3-1.domain_prediction/Predicted_Pfam_annotation_in_RXLR_effectors.tsv", sep = "\t")
```

```python
wy = pd.read_csv("../3.Clustering/3-1.domain_prediction/Predicted_WY_annotation_in_RXLR_effectors.tsv", sep = "\t")
```

```python
cluster = pd.read_csv("../3.Clustering/3-2.clustering/RXLR_cluster_domain.tsv", sep = "\t")
```

```python
pd.merge([pfam, wy, cluster], how = "inner", on = "entry")
```

```python
tmp = pd.merge(pfam, wy, how = "inner", on = ["entry", "spp"])
```

```python
total = pd.merge(tmp, cluster, how = "inner", on = ["entry", "spp"])
```

```python
total.sort_values(["str_cl", "seq_cl", "spp"]).to_csv("total.tsv", sep = "\t", index = False)
```

```python
total[["str_cl", "PFam"]].drop_duplicates().fillna("").groupby("str_cl", as_index = False).agg(["count", lambda x: '/'.join(x)]).to_csv("../3.Clustering/cluster_purity.tsv", sep = "\t", index = False)
```

```python
total[["str_cl", "PFam"]].drop_duplicates()
```

```python
total[total["str_cl"] != -1].pivot_table(index = "str_cl", columns = "seq_cl", aggfunc = "size", fill_value = 0)
```

```python
tmp = total[(total["str_cl"] != -1) & (total["str_cl"] <= 30)]
```

```python
tmp2 = tmp[tmp["seq_cl"] <= 50]
```

```python
tmp2.pivot_table(index = "str_cl", columns = "seq_cl", aggfunc = "size", fill_value = 0).to_csv("../3.Clustering/heatmap.txt", sep = "\t")
```

# data

```python
%cd ../3.Clustering/
```

```python
domains = []
for idx in total.index:
    pfam = total.loc[idx, "PFam"]
    t = total.loc[idx, "type"]

    if t == "LWY":
        domain = "LWY"
    elif t == "WY":
        domain = "WY"
    else:
        domain = pfam

    domains.append(domain)

total["domain"] = domains
```

```python
total.sort_values(["str_cl", "seq_cl", "spp"]).to_csv("total.tsv", sep = "\t", index = False)
```

```python
total.groupby("str_cl").count()
```

```python
total.dropna(subset = "PFam").groupby("str_cl").count()
```
