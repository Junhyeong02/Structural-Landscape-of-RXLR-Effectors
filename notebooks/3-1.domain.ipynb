```python
!ls -al
```

```python
with open("interproscan/RXLR_fasta_v3.json") as f:
    data = json.load(f)
```

```python
pfam_db = {}

for result in data["results"]:

    for match in result["matches"]:

        if not result["xref"][0]["name"] in pfam_db:
            pfam_db[result["xref"][0]["name"]] = []
        
        if match["signature"]["signatureLibraryRelease"]["library"] == "PFAM":
            pfam_db[result["xref"][0]["name"]].append(match["signature"]["description"])
        # elif match["signature"]["signatureLibraryRelease"]["library"] == "PANTHER":
        #     pant_db[result["xref"][0]["name"]] = match["signature"]["name"]
```

```python
for key, val in pfam_db.items():
    pfam_db[key] = '/'.join(val)
```

```python
df["PFam"] = df["entry"].map(pfam_db)
```

```python
df
```

```python
df[["entry", "spp", "PFam"]].to_csv("Predicted_Pfam_annotation_in_RXLR_effectors.tsv", sep = "\t", index = False)
```

# WY PFAM

```python
# !conda run -n homology-env hmmsearch -T 0.0 -o WY_hmmsearch.results ./wy/WY.hmm ../RXLR_db/RXLR_fasta_v3.fasta
```

```python
!ls -al
```

```python
wy_domain = {}

for query in SearchIO.parse("wy/WY_hmmsearch.results", "hmmer3-text"):
    for hit in query:
        if hit.bitscore > 0:
            wy_domain[hit.id] = len(hit.hsps)
```

```python
df["WY"] = df["entry"].map(wy_domain)
```

```python
df["WY"] = df["WY"].fillna(0)
```

# LWY PFAM

```python
# !conda run -n homology-env hmmsearch -T 0.0 -o ./wy/LWY_hmmsearch.results ./wy/LWY.hmm ../RXLR_db/RXLR_fasta_v3.fasta
```

```python
lwy_domain = {}

for query in SearchIO.parse("wy/LWY_hmmsearch.results", "hmmer3-text"):
    for hit in query:
        lwy_domain[hit.id] = sum(1 if hsp.bitscore > 30 else 0 for hsp in hit.hsps )
```

```python
df["LWY"] = df["entry"].map(lwy_domain)
```

```python
df["LWY"] = df["LWY"].fillna(0)
```

```python
df
```

```python
rxlr_type = []
```

```python
for idx in df.index:
    
    if df.loc[idx, "LWY"] != 0:
        rxlr_type.append("LWY")
    elif df.loc[idx, "WY"] != 0:
        rxlr_type.append("WY")
    else:
        rxlr_type.append("no (L)WY")
```

```python
df["type"] = rxlr_type
```
